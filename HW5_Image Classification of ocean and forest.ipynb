1. Data Preparation 

```python
# Reading the image
# Imports PIL module  
from PIL import Image 


# Python program to explain os.listdir() method  
    
# importing os module  
import os 
import matplotlib.image as mpimg
from skimage.transform import resize
import numpy as np
#Normalize the data
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
#Pandas
import pandas as pd

from skimage.io import imread 
import numpy as np 
import matplotlib.pyplot as plt 
from sklearn import svm 
from sklearn.model_selection import GridSearchCV 
from sklearn.model_selection import train_test_split 
from sklearn.metrics import accuracy_score 
from sklearn.metrics import classification_report
from sklearn import datasets, linear_model

from sklearn.datasets import make_regression
from sklearn.metrics import mean_squared_error, r2_score
import time
```

```python
#Function to read the images
def read_images(zipfile):
    # Create image array
    images = []
    # For each file in folder called zipfile
    for filename in os.listdir(zipfile):
        # Save complete address in img
        img = os.path.join(zipfile,filename)
        # Read Image in directory
        image_read = mpimg.imread(img)
        # Resize image to 16*16 into RGB 3 colors
        image_resize = resize( image_read,(16,16,3))
        #Convert image to pixels
        image_pixels = np.array(image_resize)
        # Flatten the image
        image_change = image_pixels/255
        image_flatten = image_change.flatten()
        # Append the image to image array
        images.append(image_flatten)
    return images
        
        
        
    
```

```python
#Call the function to read images and store it
test_ocean = read_images('test/ocean')
test_forest=read_images('test/forest')
train_ocean = read_images('train/ocean')
train_forest =read_images('train/forest')
```

```python
# Print the length of the sub folder
print(len(test_ocean))
print(len(test_forest))
print(len(train_ocean))
print(len(train_forest))
```

```python
#Normalize the data
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
#Scaling each file
#test_ocean
scaler.fit(test_ocean)
test_ocean = scaler.transform(test_ocean)
#test_forest
scaler.fit(test_forest)
test_forest=scaler.transform(test_forest)
#train_ocean
scaler.fit(train_ocean)
train_ocean=scaler.transform(train_ocean)
#train_forest
scaler.fit(train_forest)
train_forest=scaler.transform(train_forest)
```

```python

#dataframe 
test_ocean = pd.DataFrame(test_ocean)
test_ocean['target']=0
test_forest= pd.DataFrame(test_forest)
test_forest['target']=1
train_ocean= pd.DataFrame(train_ocean)
train_ocean['target']=0
train_forest= pd.DataFrame(train_forest)
train_forest['target']=1
```

```python
#Preparing Training Data
train_data = pd.concat([train_ocean,train_forest],axis=0)
train_data.shape
```

```python
#Prepare Test data
test_data = pd.concat([test_ocean,test_forest],axis=0)
test_data.shape
```

```python
#Train data
#input data  
X_train=train_data.iloc[:,:-1]  
#output data 
y_train=train_data.iloc[:,-1]
#input train data  
X_test=test_data.iloc[:,:-1]  
#output train data 
y_test=test_data.iloc[:,-1]
```

```python
X_train.columns = X_train.columns.astype(str)
X_test.columns = X_test.columns.astype(str)
```

2.Model Training and Optimization (10):

Train the model and use cross validation to optimize the hyperparameters. Try both the linear and rbf kernel types.
Explain your results based on your intuition

```python
# Defining the parameters grid for GridSearchCV 
param_grid={'C':[1,8,10], #Regularisation coefficient
            'gamma':[0.01,0.1,1], #Kernel coefficient
            'kernel':['linear','rbf']} #Kernel type

# Creating a support vector classifier 
svm=svm.SVC() 

# Creating a model using GridSearchCV with the parameters grid
t0 = time.time()
grid_search = GridSearchCV(svm,param_grid, scoring = 'f1',
                           return_train_score=True, 
                           cv = 5,
                           verbose = 1)
grid_search.fit(X_train,y_train)
time_gridsearch =  time.time()-t0

#Best Hyperparameters
best_parameters = grid_search.best_params_
print('The best parameter values were:', best_parameters)

```

3.Model Evaluation (3):

Train the final SVM model (using the best parameters from above) using the entire training dataset.
Report both the train and test accuracy.

```python
#Get the best SVM model
best_svm_model =grid_search.best_estimator_

#Evaluate the best model on the test set
#Train Accuracy
y_pred_train= best_svm_model.predict(X_train)
accuracy_train = accuracy_score(y_train,y_pred_train)
print('The best accuracy in train was:',accuracy_train )
#Test Accuracy
y_pred_test = best_svm_model.predict(X_test)
accuracy_test = accuracy_score(y_test,y_pred_test)
print('The best accuracy in test was:',accuracy_test )
```

Explaination:
1.Rbf
The data points for pixel are dispersed so rbf is working properly as there is no linear separation in images.
Linear kernel is used when the data is linearly separable. It means that data can be separated using a single line.
Radial basis function kernel is a general purpose kernel. It is used when we have no prior knowledge about the data. 

2.C=8
C regularization parameter represents the penalty for misclassification.
 This means if C is high, the cost of misclassification is higher, therefore in the tradeoff between margin width andclassification error, it ends up optimizing the classification error more.
When C is low, the cost of misclassification is low, therefore the model ends up maximizing the margin
more.
If C is large, we have a small margin
If C is small, we have a large margin
It gave optimum value between 1 and 10.

3.'gamma': 0.01

The gamma parameter in the RBF kernel plays a critical role in determining the shape of the decision
boundary. It controls the width of the Gaussian function used to map the input data into a higher-
dimensional space.
 A small value of gamma means that the influence of each training example is relatively large, and the
decision boundary becomes more curved or nonlinear.
It choose lowest gamma value to fit data most accurately
